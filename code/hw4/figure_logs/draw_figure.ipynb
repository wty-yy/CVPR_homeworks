```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
config = {
    "font.family": 'serif', # 衬线字体
    "figure.figsize": (14, 6),  # 图像大小
    "font.size": 20, # 字号大小
    "font.serif": ['SimSun'], # 宋体
    "mathtext.fontset": 'stix', # 渲染数学公式字体
    'axes.unicode_minus': False # 显示负号
}
plt.rcParams.update(config)
```

```python
# model_name = 'inception_resnet_v2'
# model_name1 = 'Inception-ResNet-v2'
# model_name = 'mobilenet_v2'
# model_name1 = 'MobileNet_v2'
model_name = 'efficientnet_dropout'
model_name1 = 'EfficientNet'
titles = ['准确率', '损失值']
train_acc = np.array(pd.read_csv(f'accuracy/run-{model_name}_train-tag-epoch_accuracy.csv')['Value'])
val_acc = np.array(pd.read_csv(f'accuracy/run-{model_name}_validation-tag-epoch_accuracy.csv')['Value'])
train_loss = np.array(pd.read_csv(f'loss/run-{model_name}_train-tag-epoch_loss.csv')['Value'])
val_loss = np.array(pd.read_csv(f'loss/run-{model_name}_validation-tag-epoch_loss.csv')['Value'])
```

```python
max_n = 41
train_acc = train_acc[:max_n]
val_acc = val_acc[:max_n]
train_loss = train_loss[:max_n]
val_loss= val_loss[:max_n]
```

```python
plt.subplot(1,2,1)
plt.plot(train_acc, label='训练集')
plt.plot(val_acc, label='验证集')
plt.title(titles[0])
plt.grid(linestyle = '--')
plt.legend()

plt.subplot(1,2,2)
plt.plot(train_loss, label='训练集')
plt.plot(val_loss, label='验证集')
plt.title(titles[1])
plt.grid(linestyle = '--')
plt.legend()
plt.suptitle(model_name1)
plt.savefig(f'{model_name}.png', dpi=300)
plt.show()
```
